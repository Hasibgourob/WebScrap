```python
df['revenue']= df['quantity']* df['price']
```

```python
df.groupby(df['order_date'].dt.to_period('M')).agg({'revenue':'sum'}).plot(kind = 'line', title = 'revenue month',xlabel='Month',ylabel='revenue')
```

```python
import matplotlib.pyplot as plt
month_revenue = df.groupby(df['order_date'].dt.to_period('M'))['revenue'].sum().reset_index()
month_revenue['month'] = month_revenue['order_date'].astype(str)

plt.figure(figsize=(10, 6))
plt.plot(month_revenue['month'], month_revenue['revenue'], marker='o', linestyle='-', color='blue')
plt.title('Month-wise Revenue')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
df['total_sales'] =df['price'] * df['quantity']
category_sales = df.groupby('category')['total_sales'].sum().reset_index()

# Plot Category-wise Sales
plt.figure(figsize=(10,6))
plt.bar(category_sales['category'], category_sales['total_sales'], color='skyblue')
plt.title('Category-wise Sales')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```
