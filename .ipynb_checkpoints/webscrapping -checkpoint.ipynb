```python
#Rokomari book link with title
from bs4 import BeautifulSoup
import requests
import re

url = "https://www.rokomari.com/leaderboard#fiction"

page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')


base_url = "https://www.rokomari.com"


tags = soup.find_all('a', class_='d-flex align-items-center')  

def generate_slug(title):
    slug = title.strip().replace(" ", "-")  
    slug = re.sub(r"[^\w\u0980-\u09FF-]", "", slug)  
    return slug.lower()


for tag in tags:
    book_link = tag.get('href')  # Extract href attribute
    book_title_tag = tag.find('p', class_='product-name')  # Find the book title

    if book_link and "/book/" in book_link and book_title_tag:
        book_id = book_link.split("/")[-1]  # Extract book ID from URL
        book_title = book_title_tag.text.strip()  # Get the book title
        
        book_slug = generate_slug(book_title)  # Convert title to a URL-friendly format

        full_link = f"{base_url}/book/{book_id}/{book_slug}"
        print(full_link)
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time

# Function to get the sale price from a book's webpage
def get_price(url):
    headers = {"User-Agent": "Mozilla/5.0"}
    
    try:
        response = requests.get(url, headers=headers, timeout=10)  # Timeout for stability
        
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # Extract the sale price inside <ins class="single-product-sale-price">
            price_element = soup.find("ins", class_="single-product-sale-price")
            if price_element:
                price = price_element.find("span", class_="woocommerce-Price-amount").text.strip()
                return price
            else:
                return "Price not found"
        
        return f"Failed, status {response.status_code}"
    
    except requests.exceptions.RequestException as e:
        return f"Error: {e}"

# List of book URLs to scrape
urls = [
    "https://www.wafilife.com/sokal-sondhar-dua-o-zikr/dp/257234",
    "https://www.wafilife.com/message/dp/267352",
    "https://www.wafilife.com/be-smart-with-muhammad/dp/129851",
    "https://www.wafilife.com/productive-muslim/dp/112205",
    "https://www.wafilife.com/sohih-masnun-wajifa/dp/19461",
    "https://www.wafilife.com/quraner-shobdaboli-level-1/dp/448244",
    "https://www.wafilife.com/rasulullah-s-er-namaz-1st-o-2nd-part-ekotre/dp/534398",
    "https://www.wafilife.com/ahban/dp/405613",
    "https://www.wafilife.com/revive-your-heart/dp/184612",
    "https://www.wafilife.com/the-caring-husband/dp/310057",
    "https://www.wafilife.com/halal-binodhon/dp/25442"
]

# List to store the results
book_data = []

# Scrape prices for each URL
for index, url in enumerate(urls):
    price = get_price(url)
    book_data.append({"Book URL": url, "Price": price})

    # Print progress every 2 books
    if index % 2 == 0:
        print(f"Processed {index}/{len(urls)} books...")

    time.sleep(1)  # Delay to prevent overloading the website

# Convert results to a Pandas DataFrame
df = pd.DataFrame(book_data)

# Save the DataFrame to a CSV file
df.to_csv("book_prices.csv", index=False)

print("Scraping complete. Data saved to 'book_prices.csv'.")
```

```python
##ROKOMARI price and link with Book unavailable done
import pandas as pd
import requests
from bs4 import BeautifulSoup


csv_file = 'C:/Users/DA/Desktop/Rokomari_scrap_link.csv'
links_df = pd.read_csv(csv_file)


book_data = []

# Iterate over each link in the CSV
for link in links_df['Link']:  # Assuming the column name is 'Link'
    # Send request to the URL
    response = requests.get(link)

    # If the response is successful
    if response.status_code == 200:
        # Check if the URL is the generic "book" page
        if response.url == "https://www.rokomari.com/book":
            # If it redirects to the generic page, mark it as "Book Unavailable"
            book_data.append({'Link': link, 'Price': 'Book Unavailable'})
        else:
            # Parse the HTML content if it's a valid book page
            soup = BeautifulSoup(response.content, 'html.parser')

            # Find the sell price on the page
            price = soup.find('span', class_='sell-price')  # Assuming this is the correct class for price

            if price:
                # If price is found, save the link and the price
                price = price.text.strip()
                book_data.append({'Link': link, 'Price': price})
            else:
                # If price is not found, mark as "Book Unavailable"
                book_data.append({'Link': link, 'Price': 'Book Unavailable'})

# Create a new DataFrame from the extracted data
output_df = pd.DataFrame(book_data)

# Save the output to a new CSV file
output_df.to_csv('C:/Users/DA/Desktop/Rokomari_books_prices.csv', index=False)

print("Data extraction complete!")

```
