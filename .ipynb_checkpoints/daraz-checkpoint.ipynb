```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Set Chrome options
options = Options()
options.add_argument('--headless')  # Run in headless mode if needed (browser won't open)
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Use webdriver-manager to automatically download and manage the correct version of ChromeDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Open the seller page
url = 'https://www.daraz.com.bd/shop/somoyerhat-express/?spm=a2a0e.pdp_revamp.seller.1.307a153fPa9Kq2&itemId=389078017&channelSource=pdp'
driver.get(url)

# Wait for the page to fully load
wait = WebDriverWait(driver, 10)  # Wait up to 10 seconds for elements to be visible

# Wait for the product containers to load
products = wait.until(EC.presence_of_all_elements_located((By.XPATH, '//div[contains(@data-exp-justforyou, "1")]')))

# Extract product info
product_data = []

for product in products:
    try:
        # Get the title (product name)
        title_elem = product.find_element(By.XPATH, './/span')
        title = title_elem.text.strip()
        
        # Get the price
        price_elem = product.find_element(By.XPATH, './/div[contains(@style, "font-size: 18px")]')
        price = price_elem.text.strip()

        # Get the link
        link_elem = product.find_element(By.XPATH, './/a')
        link = link_elem.get_attribute('href')

        # Append the data to the list
        product_data.append({'title': title, 'price': price, 'link': link})
    
    except Exception as e:
        print("Skipping item due to error:", e)
        continue

# Close the browser
driver.quit()

# Print or save the results
for item in product_data:
    print(item)

```

```python
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Set Chrome options
options = Options()
options.add_argument('--headless')  # Run in headless mode if needed (browser won't open)
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Use webdriver-manager to automatically download and manage the correct version of ChromeDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Open the seller page
url = 'https://www.daraz.com.bd/shop/somoyerhat-express/?spm=a2a0e.pdp_revamp.seller.1.307a153fPa9Kq2&itemId=389078017&channelSource=pdp'
driver.get(url)

# Wait for the page to fully load
wait = WebDriverWait(driver, 10)  # Wait up to 10 seconds for elements to be visible

# Wait for the product containers to load
products = wait.until(EC.presence_of_all_elements_located((By.XPATH, '//div[contains(@data-exp-justforyou, "1")]')))

# Extract product info
product_data = []

for product in products:
    try:
        # Get the title (product name)
        title_elem = product.find_element(By.XPATH, './/span')
        title = title_elem.text.strip()
        
        # Get the price
        price_elem = product.find_element(By.XPATH, './/div[contains(@style, "font-size: 18px")]')
        price = price_elem.text.strip()

        # Get the link
        link_elem = product.find_element(By.XPATH, './/a')
        link = link_elem.get_attribute('href')

        # Append the data to the list
        product_data.append({'title': title, 'price': price, 'link': link})
    
    except Exception as e:
        print("Skipping item due to error:", e)
        continue

# Close the browser
driver.quit()

# Define CSV file name
csv_file = r"C:/Users/DA/Desktop/Rokomari/daraz.csv"

# Write the data to a CSV file
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=["title", "price", "link"])
    
    # Write the header
    writer.writeheader()
    
    # Write the rows of product data
    for item in product_data:
        writer.writerow(item)

print(f"Data saved to {csv_file}")
```
