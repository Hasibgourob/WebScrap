```python
import pytesseract
from PIL import Image
import pandas as pd
import re

# Path to your image
image_path = "C:/Users/DA/Desktop/Invoice.jpg"

# Extract text from image
extracted_text = pytesseract.image_to_string(Image.open(image_path))


# Define a function to extract multi-line data more accurately
def extract_field(text, label, stop_labels=None):
    try:
        start = text.find(label) + len(label)
        if stop_labels:
            end = min([text.find(stop, start) for stop in stop_labels if text.find(stop, start) != -1])
        else:
            end = text.find("\n", start)
        return text[start:end].strip()
    except:
        return "N/A"


# Extracting data
fields = {
    "Order ID": extract_field(extracted_text, "Order ID"),
    "Invoice ID": extract_field(extracted_text, "Invoice ID:"),
    "Deliver To": extract_field(extracted_text, "Deliver To:"),
    "Delivery Address": extract_field(extracted_text, "Delivery Address:"),
    "Bill To": extract_field(extracted_text, "Bill To:"),
    "Billing Address": extract_field(extracted_text, "Billing Address:"),
    "SellerCenter": "N/A",
    "Order Date": extract_field(extracted_text, "Order Date:"),
    "Paid by": extract_field(extracted_text, "Paid by:"),
    "Phone": extract_field(extracted_text, "Phone:"),
}

# Extract Product Name by capturing multiple lines until the price part starts
product_pattern = re.compile(
    r"Ordered Items:\s*([\s\S]*?)\d{9,}"
)
product_match = product_pattern.search(extracted_text)
fields["Product Name"] = product_match.group(1).replace("\n", " ").strip() if product_match else "N/A"

# Extract SKU, Size, Price, Total, etc.
fields["Paid price"] = extract_field(extracted_text, "price", ["Total"])
fields["Shop SKU"] = re.search(r"(\d{9,}_BD)", extracted_text).group(1) if re.search(r"(\d{9,}_BD)", extracted_text) else "N/A"
fields["Item"] = "1"
fields["Seller SKU"] = extract_field(extracted_text, "Seller SKU")
fields["Size"] = re.search(r"Size\s*:\s*([A-Za-z0-9]+)", extracted_text).group(1) if re.search(r"Size\s*:\s*([A-Za-z0-9]+)", extracted_text) else "N/A"
fields["Price"] = extract_field(extracted_text, "Price", ["Subtotal"])
fields["Total"] = extract_field(extracted_text, "Total")

# Convert to DataFrame
df = pd.DataFrame([fields])

# Save to CSV
df.to_csv("invoice_from_image_1.csv", index=False)

print("âœ… Invoice data extracted and saved to CSV!")
```

```python
import pytesseract
from PIL import Image
import pandas as pd
import re

# Load the image
image_path = 
extracted_text = pytesseract.image_to_string(Image.open(image_path))

# Define a function to extract fields with more flexible regex
def extract_field(text, label, stop_label=None):
    start = text.find(label)
    if start == -1:
        return "N/A"
    start += len(label)
    if stop_label:
        end = text.find(stop_label, start)
        if end == -1:
            end = len(text)
    else:
        end = text.find("\n", start)
    return text[start:end].replace("\n", " ").strip()

# Improved phone number extraction to handle scientific notation (e.g., 8.80181E+12)
def extract_phone(text):
    phone_match = re.search(r"Phone[:\s]*(\d{10,15})", text)
    if phone_match:
        return phone_match.group(1)
    return "N/A"

# Improved price extraction: handle cases like scientific notation, missing decimals
def extract_price(text):
    price_pattern = re.findall(r"(\d{1,4}\.\d{2}|\d{1,5}(?:E[\+\-]?\d{1,3})?)", text)
    return price_pattern

# Extract invoice header data
data = {
    "Invoice ID": extract_field(extracted_text, "Invoice ID:", "Paid by"),
    "Deliver To": extract_field(extracted_text, "Deliver To:", "Phone"),
    "Delivery Address": extract_field(extracted_text, "Delivery Address:", "Bill To"),
    "Bill To": extract_field(extracted_text, "Bill To:", "Phone"),
    "Billing Address": extract_field(extracted_text, "Billing Address:", "Ordered Items"),
    "SellerCenter": "N/A",  # Placeholder for SellerCenter info
    "Order Date": extract_field(extracted_text, "Order Date:", "Invoice ID"),
    "Paid by": extract_field(extracted_text, "Paid by:", "Deliver To"),
    "Phone": extract_phone(extracted_text),
}

# Extract product details and clean the product name
product_match = re.search(r"Ordered Items:\s*(.*?)\s*Subtotal", extracted_text, re.DOTALL)
data["Product Name"] = product_match.group(1).replace("\n", " ").strip() if product_match else "N/A"

# Extract price-related fields (handling scientific notation and missing decimals)
price_pattern = extract_price(extracted_text)

# Handle missing values and add "N/A" for missing data
data["Paid price"] = price_pattern[0] if len(price_pattern) > 0 else "N/A"
data["Price"] = price_pattern[1] if len(price_pattern) > 1 else "N/A"
data["Total"] = price_pattern[-1] if len(price_pattern) > 2 else "N/A"

# Extract SKU, size, and item count
data["Shop SKU"] = re.search(r"(\d{9,}_BD)", extracted_text).group(1) if re.search(r"(\d{9,}_BD)", extracted_text) else "N/A"
data["Item"] = "1"
data["Seller SKU"] = extract_field(extracted_text, "Seller SKU", "Size")
data["Size"] = re.search(r"Size\s*:\s*([A-Za-z0-9]+)", extracted_text).group(1) if re.search(r"Size\s*:\s*([A-Za-z0-9]+)", extracted_text) else "N/A"

# Convert to DataFrame and save as CSV
df = pd.DataFrame([data])
df.to_csv("clean_invoice_data.csv", index=False)

print("âœ… Invoice data extracted and saved to CSV!")
```
